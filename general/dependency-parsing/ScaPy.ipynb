```python
#python -m spacy download en_core_web_sm
import spacy

nlp = spacy.load('en_core_web_sm')
sentence = 'An old man is going to the river bank.'
doc = nlp(sentence)

print("{:<15} | {:<8} | {:<15} | {:<20}".format('Token', 'Relation', 'Head', 'Children'))
print("-" * 70)

for token in doc:
    print("{:<15} | {:8} | {:<15} | {:<20}".format(str(token.text), str(token.dep_), str(token.head), str([child for child in token.children])))
```

```python
import spacy.displacy

spacy.displacy.render(doc, style='dep', jupyter=True, options={'distance': 130})
```
